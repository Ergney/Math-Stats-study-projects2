Задача 2

Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Так как распределение является нормальным, а для номрально распределния среднее значение равно мат ожиданию a = 174, 
$\sigma$ = 8

```python
x = stats.norm(loc=174, scale=8)
1 - x.cdf(182)
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см: {1 - x.cdf(182)}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см: {1 - x.cdf(190)}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см: {x.cdf(190) - x.cdf(166)}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см: {x.cdf(182) - x.cdf(166)}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см: {x.cdf(190) - x.cdf(158)}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см: {x.cdf(150) + (1 - x.cdf(190))}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см: {x.cdf(150) + (1 - x.cdf(198))}')
print(f'вероятность того, что случайным образом выбранный взрослый человек имеет рост ниже 166 см: {x.cdf(166)}')
```
